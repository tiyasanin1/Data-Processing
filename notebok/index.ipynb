Kolom-kolom dari dataset 'diamonds.csv' memiliki definisi sebagai berikut:

- price: price in US dollars ( 326âˆ’âˆ’ 18,823)
- carat: weight of the diamond (0.2--5.01)
- cut: quality of the cut (Fair, Good, Very Good, Premium, Ideal)
- color: diamond colour, from J (worst) to D (best)
- clarity: a measurement of how clear the diamond is (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))
- x: length in mm
- y: width in mm
- z: depth in mm
- depth: total depth percentage = z / mean(x, y) = 2 * z / (x + y)
- table: width of top of diamond relative to widest point

```python
df['carat'].fillna(df['carat'].median(),inplace=True)
df
```

```python
df['clarity'].unique()
```

```python
#change clarity from categorical to numeric
change_clarity={
    'I1':1,
    'SI2':2, 
    'SI1':3,
    'VS2':4, 
    'VS1':5, 
    'VVS2':6, 
    'VVS1':7, 
    'IF':8
}
df['clarity']=df['clarity'].map(change_clarity)
df 
```

```python
#outlier identification 

# X boxplot
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
sns.boxplot(x=df['x'])
plt.title('boxplot X')

# y boxplot
plt.figure(figsize=(15,5))
plt.subplot(1,3,2)
sns.boxplot(x=df['y'])
plt.title('boxplot y')

# z boxplot
plt.figure(figsize=(15,5))
plt.subplot(1,3,3)
sns.boxplot(x=df['z'])
plt.title('boxplot z')
```

```python
outlier={}
for column in ['x','y','z']:
    q1=df[column].quantile(0.25)
    q3=df[column].quantile(0.75)
    q2=q3-q1
    batas_bawah=q1-1.5*q2
    batas_atas=q3+1.5*q2
    outlier[column]=df[(df[column]<batas_bawah)|(df[column]>batas_atas)]
#See outlier in x,y,z column
all_outlier=pd.concat([outlier['x'],outlier['y'],outlier['z']])
all_outlier
```
